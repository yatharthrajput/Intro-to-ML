# BootTrapping 

```python
sample=df_sample['travel_time'].values
```

```python
print(sample,len(sample))
```

```python
size=len(sample)
size
```

```python
r=100
bs_samples=[]
```

```python
for i in range(r):
    a=np.random.choice(sample,size=size).mean()
    bs_samples.append(a)
    
```

```python
bs_samples=np.array(bs_samples)
bs_samples
```

```python
sns.histplot(bs_samples,bins=10,kde=True)
```
